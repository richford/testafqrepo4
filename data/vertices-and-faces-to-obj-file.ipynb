```python
from glob import glob
```

```python
dirname = 'alpha_shapes'
vertex_ext = '.vertices'
facets_ext = '.alpha_shape.facets'
filenames = glob(dirname + '/' + '*' + vertex_ext)

with open('alpha_shapes.OBJ', 'w') as obj_file:
    for fn in filenames:
        bundle_name = fn.replace(dirname + '/', '').replace(vertex_ext, '').replace('_', '-')

        obj_file.write('o ' + bundle_name + '\n')

        with open(fn, 'r') as f:
            f.readline()
            for line in f:
                obj_file.write('v ' + line)

        facets_fn = fn.replace(vertex_ext, facets_ext)
        with open(facets_fn, 'r') as f:
            for line in f:
                obj_file.write('f ' + line)
```
